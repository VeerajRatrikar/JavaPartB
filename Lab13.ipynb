```python
import pandas as pd
import matplotlib.pyplot as plt
data  = pd.read_csv('FremontBridge.csv',index_col='Date',parse_dates=True)
#index column is a time series object
data.columns=['Total','East','West']
data.head()
```

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
print(by_weekday)
by_weekday.index=['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
by_weekday.plot(style=[':','--','-'])
```

```python
# Group by month and calculate the mean for each month
by_month = data.groupby(data.index.month).mean()
print(by_month)

# Rename the index to month names
by_month.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Plot with different line styles for each month
#styles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--', '-.', ':']  # 12 styles for 12 months
by_month.plot(style=[':','--','-'])
```

```python
# Group by month and calculate the mean for each month
by_year = data.groupby(data.index.year).mean()
print(by_month)

# Rename the index to month names
by_year.index = ['2019','2020','2021']

# Plot with different line styles for each month
#styles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--', '-.', ':']  # 12 styles for 12 months
by_year.plot(style=[':','--','-'])
```
