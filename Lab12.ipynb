```python
#Write a NumPy program to create a structured array from given student name,height,class and their data types with some 
#missing values and show the data.Fill out the missing values and output the data.
import numpy as np

# Define the data type with a float for height and float for class
data_type = [('name', 'S15'), ('class', float), ('height', float)]

# Create a structured numpy array with 4 students, with some missing values
students = np.zeros(4, dtype=data_type)

# Assign data with missing values (represented by np.nan for height and class)
students['name'] = ['amiths', 'dhanush', 'hemanth', 'hit']
students['class'] = [5, np.nan, 5, 5]  # Using np.nan for missing class for 'dhanush'
students['height'] = [48.5, 53, np.nan, 40]  # Missing height for 'hemanth'

# Output the original data with missing values
print("Original Data (with missing values):")
print(students)

# Fill missing values:
# - For 'class', replace missing values (NaN) with the most common class (5).
# - For 'height', replace missing values with the mean height.

# Fill missing 'class' with 5 (most common class) where class is NaN
students['class'] = np.where(np.isnan(students['class']), 5, students['class'])

# Fill missing 'height' with the mean of the existing height values
mean_height = np.nanmean(students['height'])
students['height'] = np.where(np.isnan(students['height']), mean_height, students['height'])

# Output the data after filling missing values
print("\nData after filling missing values:")
print(students)
```
